```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("projeto_aplicado.db") 

query = """
SELECT 
    e.municipio,
    e.trimestre,
    e.`Nivel de Instrução` AS nivel_instrucao,
    e.`quantidade de escolaridade` AS quantidade_escolaridade,
    p.`posição de ocupação`,
    p.`quantidade de posição de ocupação`,
    c.`Condição de ocupação`,
    c.`quantidade de condição de ocupação`,
    i.Grupo AS grupo_etario,
    i.`quantidade de idade`
FROM 
    tabela_escolaridade_atualizado e
LEFT JOIN 
    tabela_posicao_atualizado p
ON 
    e.municipio = p.municipio AND e.trimestre = p.trimestre
LEFT JOIN 
    tabela_condicao_atualizado c
ON 
    e.municipio = c.municipio AND e.trimestre = c.trimestre
LEFT JOIN 
    tabela_idade_atualizado i
ON 
    e.municipio = i.municipio AND e.trimestre = i.trimestre;
"""

df_merged = pd.read_sql(query, conn)

print("Dados consolidados:")
print(df_merged.head())

conn.close()
```

```python
def recomendar_programas_com_porcentagem(municipio, df_consolidado):

    dados_municipio = df_consolidado[df_consolidado["municipio"] == municipio]
    
    if dados_municipio.empty:
        print(f"Nenhum dado disponível para o município: {municipio}")
        return

    print(f"\nPorcentagem de pessoas que podem se beneficiar dos programas no município: {municipio}")

    total_pessoas = dados_municipio["quantidade de idade"].sum()
    
    if total_pessoas == 0:
        print("Não há registros suficientes para calcular porcentagens.")
        return

    recomendacoes = []

    if (dados_municipio["grupo_etario"].str.contains("25 a 39 anos").any()):
        beneficiarios = dados_municipio.loc[dados_municipio["grupo_etario"] == "25 a 39 anos", "quantidade de idade"].sum()
        porcentagem = (beneficiarios / total_pessoas) * 100
        recomendacoes.append(f"- Minha Casa Minha Vida: {porcentagem:.2f}% das pessoas estão na faixa etária beneficiada.")

    if (dados_municipio["grupo_etario"].str.contains("40 a 59 anos").any() and
        dados_municipio["Condição de ocupação"].str.contains("Fora da força de trabalho").any()):
        beneficiarios = dados_municipio.loc[dados_municipio["grupo_etario"] == "40 a 59 anos", "quantidade de idade"].sum()
        porcentagem = (beneficiarios / total_pessoas) * 100
        recomendacoes.append(f"- Auxílio Emergencial: {porcentagem:.2f}% das pessoas estão fora da força de trabalho.")

    if (dados_municipio["grupo_etario"].str.contains("60 anos ou mais").any()):
        beneficiarios = dados_municipio.loc[dados_municipio["grupo_etario"] == "60 anos ou mais", "quantidade de idade"].sum()
        porcentagem = (beneficiarios / total_pessoas) * 100
        recomendacoes.append(f"- BPC: {porcentagem:.2f}% das pessoas são idosos que podem se beneficiar.")

    if (dados_municipio["grupo_etario"].str.contains("18 a 24 anos").any()):
        beneficiarios = dados_municipio.loc[dados_municipio["grupo_etario"] == "18 a 24 anos", "quantidade de idade"].sum()
        porcentagem = (beneficiarios / total_pessoas) * 100
        recomendacoes.append(f"- Acessuas Trabalho: {porcentagem:.2f}% das pessoas estão na faixa jovem elegível para qualificação.")

    for recomendacao in recomendacoes:
        print(recomendacao)
```
